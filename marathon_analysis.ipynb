```python
import matplotlib.pyplot as plt

# Liste des kilomètres par semaine depuis début août
distances = [
    30, 40, 140, 160, 170, 160, 165, 165, 160, 160, 160, 160, 160, 160,
    90, 105, 80, 80, 130, 190, 140, 190, 200, 130, 150, 180, 185, 200,
    160, 200, 210, 190, 100, 170, 210, 200, 55, 0
]

# Événements spéciaux avec leurs annotations
events = {
    4: "Prépa du semi de Boulogne en contrôle (pas de séances plus rapide que 3:05/km), \n                       je fais l'autiste à 160km/semaine",
    16: "Semi Boulogne 1h05:06 (passage au 10km en 29:50)",
    23: "10km Nice 29:26",
    27: "Première spé marathon: 5km 14:50 + 10km 30:30 (matin)/ 10x1km R1 (aprem)",
    30: "Double seuils 3x10', 30x400 / 5x6', 25x400; sortie longue : 40km @3:20",
    31: "trois double seuil même format",
    32: "15x1km/1km (3:00/3:30) / Semi Paris 1h04:40",
    34: "25x400 / 15x1000 / 10km + 4x5km",
    35: "25x400 / 12x1000 / 20km tempo",
    37: "Fin du game"
}

weeks = list(range(1, len(distances) + 1))

# Créer une figure plus grande
plt.figure(figsize=(24, 12))
plt.plot(weeks, distances, marker='o', linestyle='-', color='blue', label='Km hebdo')
plt.title("Kilométrage hebdomadaire depuis début août", fontsize=30)
plt.xlabel("Semaine", fontsize=22)
plt.ylabel("Kilomètres", fontsize=22)

# Ajouter des marqueurs rouges pour les événements
event_weeks = list(events.keys())
event_labels = list(events.values())
event_kms = [distances[i] for i in event_weeks]

plt.scatter([weeks[i] for i in event_weeks], event_kms, color='red', s=150, label='Événements clés', zorder=5)

# Créer une légende textuelle avec les événements listés
legend_entries = [f"Semaine {weeks[i]} : {label}" for i, label in zip(event_weeks, event_labels)]
legend_text = "\n".join(legend_entries)

# Afficher la légende à droite de la figure
plt.gcf().text(1.02, 0.5, legend_text, fontsize=20, va='center', ha='left', bbox=dict(facecolor='white', edgecolor='black'))

# Finalisation
plt.grid(True)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Laisser de la place pour la légende
plt.show()
```
